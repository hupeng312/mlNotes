```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
```

```python
cosy = np.cos(x)
```

```python
plt.plot(x, cosy, color='red', linestyle='--', label='cos(x)')
plt.plot(x, y, label='sin(x)')
plt.xlim(-5,20)
#plt.ylim(0,10)
plt.axis([-4,10,-3,10])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Welcome come!')
plt.show()
```

```python
plt.scatter(x, y)
plt.scatter(x, cosy)
plt.show()
```
